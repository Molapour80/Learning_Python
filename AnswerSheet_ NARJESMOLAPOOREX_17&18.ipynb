    <p>
        Using four different examples, demonstrate all ofthe inheritance types that you have learned.
        </p> 
<li>Single Inheritance</li>
<li>Multiple Inheritance</li>
<li>Multilevel Inheritance</li> 
<li>Hierarchical Inheritance</li>

```python
#Write code here:
#Single Inheritance
class Animal:  
    def speak(self):
        return "Animal "

class Dog(Animal):  
    def bark(self):
        return "Dog"

dog = Dog()
print(dog.speak())  
print(dog.bark())   


#Multiple Inheritance
class Flyer:  
    def fly(self):
        return "Can flyingggg"

class Swimmer:  
    def swim(self):
        return "Can swiming"

class Duck(Flyer, Swimmer):  
    def quack(self):
        return "Duck"

duck = Duck()
print(duck.fly())  
print(duck.swim())  
print(duck.quack())  

#Multilevel Inheritance

class Animal:  
    def eat(self):
        return "Animal"

class Mammal(Animal): 
    def walk(self):
        return " walks"

class Dog(Mammal):  
    def bark(self):
        return "Dog "

dog = Dog()
print(dog.eat())  
print(dog.walk())  
print(dog.bark()) 

#Hierarchical Inheritance

class Vehicle:  
    def start(self):
        return "starts"

class Car(Vehicle):  
    def drive(self):
        return " drives"

class Bike(Vehicle):  
    def ride(self):
        return "rides"

car = Car()
bike = Bike()

print(car.start())  
print(car.drive())  
print(bike.start())  
print(bike.ride())  
```

    <h4>Question 01 (Abstraction)(180):</h4>
    <p>
     Develop a library management system in Python with an abstract base class and specific
subclasses using abstraction and polymorphism.
    </p>
    <p>Abstract Base Class - LibraryItem:</p>
    <li>Attributes: ID,title, author.</li>
<li>Abstract Method - display_details: Declare withoutimplementation. Subclasses will provide specific
details.</li>
<p>Subclass - Book:<p>
<li>Additional Attributes: pages, genre</li>
<li>display_details: Implementto print book's ID,title, author, pages, and genre<li>
<p>Subclass - Magazine:</p>
<li>Additional Attributes: issue_number, publication_date.</li>
<li>display_details: Implementto print magazine's ID,title, issue number, and publication date.</li>
<p>Subclass - Article:<p>
<li>Additional Attributes: publication_date, summary.</li>
<li>display_details: Implementto print article's ID,title, publication date, and summary.</li>
</div>

```python
#Write code here:
from abc import ABC, abstractmethod

# Abstract Base Class
class LibraryItem(ABC):
    def __init__(self, item_id, title, author):
        self.item_id = item_id
        self.title = title
        self.author = author

    @abstractmethod
    def display_details(self):
        pass


# Subclass for Book
class Book(LibraryItem):
    def __init__(self, item_id, title, author, pages, genre):
        super().__init__(item_id, title, author)
        self.pages = pages
        self.genre = genre

    def display_details(self):
        return (f"Book ID: {self.item_id}\n"
                f"Title: {self.title}\n"
                f"Author: {self.author}\n"
                f"Pages: {self.pages}\n"
                f"Genre: {self.genre}")


# Subclass for Magazine
class Magazine(LibraryItem):
    def __init__(self, item_id, title, author, issue_number, publication_date):
        super().__init__(item_id, title, author)
        self.issue_number = issue_number
        self.publication_date = publication_date

    def display_details(self):
        return (f"Magazine ID: {self.item_id}\n"
                f"Title: {self.title}\n"
                f"Author: {self.author}\n"
                f"Issue Number: {self.issue_number}\n"
                f"Publication Date: {self.publication_date}")


# Subclass for Article
class Article(LibraryItem):
    def __init__(self, item_id, title, author, publication_date, summary):
        super().__init__(item_id, title, author)
        self.publication_date = publication_date
        self.summary = summary

    def display_details(self):
        return (f"Article ID: {self.item_id}\n"
                f"Title: {self.title}\n"
                f"Author: {self.author}\n"
                f"Publication Date: {self.publication_date}\n"
                f"Summary: {self.summary}")


# Example usage
if __name__ == "__main__":
    book = Book(1, "1984", "python", 328, "python")
    magazine = Magazine(2, "National Geographic", "Various", 2023, "October 2023")
    article = Article(3, "AI and Society", "Jane Doe", "2023-10-15", "An exploration of AI's impact on society.")

    for item in (book, magazine, article):
        print(item.display_details())
        print("-" * 40)
```
